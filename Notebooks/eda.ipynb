## Dataset: Adult (UCI ML Repository)
El dataset Adult, también conocido como Census Income Dataset, es un conjunto de datos ampliamente utilizado en análisis estadístico y aprendizaje automático, cuyo objetivo principal es predecir si una persona percibe ingresos anuales superiores a 50.000 dólares estadounidenses a partir de información demográfica y laboral obtenida del censo de Estados Unidos de 1994. Este conjunto de datos contiene un total de 48.842 observaciones y 15 variables en total, de las cuales 14 corresponden a variables predictoras relacionadas con características como edad, nivel educativo, ocupación, estado civil y horas trabajadas por semana y una variable objetivo denominada income, que clasifica el nivel de ingresos en dos categorías: menor o igual a 50.000 USD y mayor a 50.000 USD. 

+ *age* : Edad de la persona en años
+ *workclass* : Tipo de empleo o sector laboral en el que trabaja la persona (por ejemplo: privado, gobierno, trabajo por cuenta propia).
+ *fnlwgt* : Peso muestral final asignado por la Oficina del Censo.
+ *education* : Nivel educativo más alto alcanzado (por ejemplo: bachillerato, licenciatura, maestría).
+ *education-num* : Número de años de educación formal completados.
+ *marital-status* : Estado civil de la persona (soltero, casado, divorciado, etc.).
+ *occupation* : Tipo de ocupación o profesión desempeñada (por ejemplo: técnico, ventas, servicios, gerencial).
+ *relationship* : Relación dentro del hogar (esposo/a, hijo/a, jefe de hogar, etc.).
+ *race* : Grupo racial al que pertenece la persona.
+ *sex* :Sexo de la persona (masculino o femenino).
+ *capital-gain* : Ganancias de capital reportadas durante el año (por ejemplo, inversiones o venta de activos).
+ *capital-loss* : Pérdidas de capital registradas durante el año.
+ *hours-per-week* : Número de horas trabajadas por semana.
+ *native-country* : Pais de origen de la persona
+ *income* : Variable que indica si el ingreso anual de la persona es superior o inferior a 50.000 USD

## Imputación de los datos faltantes

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import missingno as msno

from ucimlrepo import fetch_ucirepo

# cargar datos
adult = fetch_ucirepo(id=2)
df = pd.concat([adult.data.features, adult.data.targets], axis=1)

# estilo profesional
sns.set_theme(style="whitegrid", palette="deep")
plt.rcParams['figure.figsize'] = (10,6)
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 12



colors = ["#2E86C1", "#E74C3C"]

ax = sns.countplot(
    data=df,
    x="income",
    palette=colors
)

plt.title("Distribución del nivel de ingresos", fontweight="bold")
plt.xlabel("Nivel de ingreso")
plt.ylabel("Frecuencia")

for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x()+0.3, p.get_height()+500))

plt.show()
```
